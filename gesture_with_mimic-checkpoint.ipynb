```python
import cv2
import mediapipe as mp
import sounddevice as sd
import soundfile as sf
import numpy as np
import sys
from PyQt5.QtWidgets import QApplication, QLabel, QWidget, QPushButton, QScrollArea, QVBoxLayout
from PyQt5.QtGui import QImage, QPixmap, QPainter, QColor, QPen, QFont
from PyQt5.QtCore import QTimer, Qt, QPoint
import time
import openai
import speech_recognition as sr
from pydub import AudioSegment
import openai

# OpenAI API Key
openai.api_key = "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"

mp_face_mesh = mp.solutions.face_mesh

class CameraFeed(QWidget):
    def __init__(self):
        super().__init__()

        self.setWindowTitle("Audio Recording Interface")
        self.setGeometry(100, 100, 1280, 720)

        # ArayÃ¼z alanÄ±
        self.interface_label = QLabel(self)
        self.interface_label.setFixedSize(1280, 400)
        self.interface_label.setStyleSheet("background-color: lightgray; border: 1px solid black;")
   
        # Talimatlar QLabel'i
        self.instructions_label = QLabel(self)
        self.instructions_label.setGeometry(0, 400, 1280, 600)
        self.instructions_label.setStyleSheet("font-size: 14px; color: white; background-color: black; padding: 10px;")
        self.instructions_label.setAlignment(Qt.AlignTop | Qt.AlignLeft)
        self.instructions_label.setText("""
        Instructions:
        1. Face Tracking with Camera: 
            The app controls the cursor on the screen by following the nose movements. Move your head to move the cursor.
        2. Microphone Button:
            Move the cursor over the green microphone icon.
            When you hover over the microphone icon for 3 seconds, audio recording starts.
            If you refocus on the icon while audio recording is in progress, the recording stops.
        3. Manage Voice Recording:
            When your voice recording is complete, the system automatically analyses the recording.
            After the audio analysis is completed, the content is transferred to the GPT interface and the corresponding results are displayed.
        4. Other Features of the Interface:
            You can press Q to close the application.
        5. Important Notes:
            The application performs best in a quiet environment.
            Make sure to focus on the microphone icon; the operation only takes place on the icon.
            As the camera is constantly on, make sure your background is appropriate.
        """)

        # Ä°mleÃ§ pozisyonu bilgisi
        self.cursor_label = QLabel(self)
        self.cursor_label.setText("Burun Ä°mleci KoordinatlarÄ±: X: 640, Y: 200")
        self.cursor_label.setStyleSheet("font-size: 16px; color: green;")
        self.cursor_label.setGeometry(10, self.interface_label.y() - 30, 400, 30)

        # Ãœst mesaj alanÄ±
        self.top_message_label = QLabel(self)
        self.top_message_label.setText("")
        self.top_message_label.setStyleSheet("font-size: 24px; color: blue; font-weight: bold;")
        self.top_message_label.setGeometry(400, self.interface_label.y() + 50, 500, 50)
        self.top_message_label.setAlignment(Qt.AlignCenter)

        # Alt mesaj alanÄ±
        self.bottom_message_label = QLabel(self)
        self.bottom_message_label.setText("")
        self.bottom_message_label.setStyleSheet("font-size: 20px; color: green; font-weight: bold;")
        self.bottom_message_label.setGeometry(400, self.interface_label.y() + 120, 500, 50)
        self.bottom_message_label.setAlignment(Qt.AlignCenter)

        # Mikrofon butonu
        self.mic_button = {
            "x": (self.interface_label.width() - 150) // 2, 
            "y": (self.interface_label.height() - 150) // 2,  
            "width": 150,
            "height": 150,
            "color": QColor(0, 200, 0)
        }

        # Timer
        self.timer = QTimer(self)
        self.timer.timeout.connect(self.update_frame)

        # Mediapipe FaceMesh
        self.face_mesh = mp_face_mesh.FaceMesh(refine_landmarks=True, min_detection_confidence=0.7, min_tracking_confidence=0.7)

        # Kamera
        self.cap = cv2.VideoCapture(0)
        self.cap.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
        self.cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)
        self.timer.start(20)

        # Hassasiyet Ã§arpanlarÄ±
        self.x_sensitivity = 3
        self.y_sensitivity = 5

        # Kamera kÄ±rpma oranlarÄ±
        self.crop_x = 0.25
        self.crop_y = 0.15

        # Ä°mleÃ§ pozisyonu
        self.cursor_x = self.interface_label.width() // 2
        self.cursor_y = self.interface_label.height() // 2

        # GÃ¶z kÄ±rpma
        self.left_eye_closed = False
        self.right_eye_closed = False

        # Ses kaydÄ±
        self.recording = False
        self.audio_frames = []
        self.samplerate = 48000
        self.channels = 1

        # KayÄ±t sÃ¼resi
        self.start_time = None
        self.max_duration = 60  # Maksimum kayÄ±t sÃ¼resi (saniye)

        # TÄ±klama iÅŸlemi
        self.hover_start_time = None
        self.hover_duration = 3  # 3 saniye

    def update_frame(self):
        ret, frame = self.cap.read()
        if not ret:
            return

        # GÃ¶rÃ¼ntÃ¼yÃ¼ yatay eksende Ã§evir
        frame = cv2.flip(frame, 1)

        # GÃ¶rÃ¼ntÃ¼yÃ¼ kÄ±rp
        h, w, c = frame.shape
        x_start = int(w * self.crop_x)
        x_end = w - x_start
        y_start = int(h * self.crop_y)
        y_end = h - y_start
        cropped_frame = frame[y_start:y_end, x_start:x_end]

        # BGR -> RGB
        rgb_frame = cv2.cvtColor(cropped_frame, cv2.COLOR_BGR2RGB)

        # Mediapipe ile yÃ¼z analizi
        results = self.face_mesh.process(rgb_frame)

        if results.multi_face_landmarks:
            for face_landmarks in results.multi_face_landmarks:
                # GÃ¶z kÄ±rpma durumunu kontrol et
                self.left_eye_closed = self.eye_closed(face_landmarks, [159, 145])
                self.right_eye_closed = self.eye_closed(face_landmarks, [386, 374])

                if self.left_eye_closed and self.right_eye_closed:
                    self.toggle_recording()

                # Burun imleci pozisyonunu al
                nose_tip = face_landmarks.landmark[1]
                new_cursor_x = int(nose_tip.x * self.interface_label.width())
                new_cursor_y = int(nose_tip.y * 400)

                if new_cursor_x != self.cursor_x or new_cursor_y != self.cursor_y:
                    self.cursor_x = int((new_cursor_x - self.interface_label.width() // 2) * self.x_sensitivity + self.interface_label.width() // 2)
                    self.cursor_y = int((new_cursor_y - self.interface_label.height() // 2) * self.y_sensitivity + self.interface_label.height() // 2)

                # SÄ±nÄ±rlandÄ±rma
                self.cursor_x = max(0, min(self.cursor_x, self.interface_label.width()))
                self.cursor_y = max(0, min(self.cursor_y, 400))

                # TÄ±klama iÅŸlemi kontrolÃ¼
                self.check_hover()

                # Ä°mleÃ§ pozisyonu bilgisini gÃ¼ncelle
                self.cursor_label.setText(f"Burun Ä°mleci KoordinatlarÄ±: X: {self.cursor_x}, Y: {self.cursor_y}")
                self.repaint_interface_label(self.cursor_x, self.cursor_y)

    def toggle_recording(self):
        if self.is_cursor_on_button(self.mic_button):
            if not self.recording:
                self.start_recording()
            else:
                self.stop_recording()

    def eye_closed(self, landmarks, indices):
        vertical = abs(landmarks.landmark[indices[0]].y - landmarks.landmark[indices[1]].y)
        horizontal = abs(landmarks.landmark[indices[0]].x - landmarks.landmark[indices[1]].x)
        return vertical / horizontal < 0.2

    def start_recording(self):
        self.start_time = time.time()
        self.top_message_label.setText("Audio Recording Initiated")
        self.audio_frames = sd.rec(int(self.max_duration * self.samplerate), samplerate=self.samplerate, channels=self.channels, dtype='int16')
        self.recording = True

    def stop_recording(self):
        if self.recording:
            sd.stop()
            elapsed_time = time.time() - self.start_time
            elapsed_time = min(elapsed_time, self.max_duration)
            self.recording = False
            self.top_message_label.setText("Voice Recording Complete")

            # Kaydedilen ses dosyasÄ±nÄ± kaydet
            output_file = "output.wav"
            sf.write(output_file, self.audio_frames[:int(elapsed_time * self.samplerate)], self.samplerate)
            QTimer.singleShot(2000, lambda: self.top_message_label.setText(""))

            # Ses dosyasÄ±nÄ± iÅŸleyip GPT'ye gÃ¶nder
            self.process_audio_with_gpt(output_file)

    def process_audio_with_gpt(self, audio_file):
        try:
            text = self.audio_to_text(audio_file)
            if text:
                # Metin alÄ±ndÄ±ÄŸÄ±nda diÄŸer arayÃ¼zÃ¼ aÃ§
                self.open_gpt_grid_app(text)
            else:
                self.top_message_label.setText("Metin DÃ¶nÃ¼ÅŸÃ¼mÃ¼ BaÅŸarÄ±sÄ±z!")
        except Exception as e:
            self.top_message_label.setText(f"Hata: {str(e)}")

    def audio_to_text(self, audio_file):
        recognizer = sr.Recognizer()
        audio = AudioSegment.from_wav(audio_file)
        wav_file = "temp.wav"
        audio.export(wav_file, format="wav")
        with sr.AudioFile(wav_file) as source:
            audio_data = recognizer.record(source)
            return recognizer.recognize_google(audio_data, language="tr-TR,en-US")

    def query_gpt(self, prompt_text):
        response = openai.ChatCompletion.create(
            model="gpt-3.5-turbo",
            messages=[
                {"role": "system", "content": "You are a helpful assistant."},
                {"role": "user", "content": prompt_text}
            ]
        )
        return response['choices'][0]['message']['content']

    def is_cursor_on_button(self, button):
        return (button["x"] <= self.cursor_x <= button["x"] + button["width"] and
                button["y"] <= self.cursor_y <= button["y"] + button["height"])

    def check_hover(self):
        if self.is_cursor_on_button(self.mic_button):
            if self.hover_start_time is None:
                self.hover_start_time = time.time()
            elif time.time() - self.hover_start_time >= self.hover_duration:
                self.toggle_recording()
                self.hover_start_time = None
        else:
            self.hover_start_time = None

    def repaint_interface_label(self, x, y):
        pixmap = QPixmap(self.interface_label.size())
        pixmap.fill(Qt.transparent)

        painter = QPainter(pixmap)

        # Mikrofon butonunu Ã§iz
        painter.setBrush(self.mic_button["color"])
        painter.drawEllipse(self.mic_button["x"], self.mic_button["y"], self.mic_button["width"], self.mic_button["height"])

        # Mikrofon sembolÃ¼
        painter.setBrush(QColor(255, 255, 255))
        painter.setFont(QFont("Arial", 50))
        painter.drawText(self.mic_button["x"] + 51, self.mic_button["y"] + 93, "ðŸŽ¤")

        # Ä°mleÃ§ Ã§emberi Ã§iz
        painter.setBrush(QColor(255, 0, 0))
        painter.setPen(Qt.NoPen)
        painter.drawEllipse(QPoint(x, y), 20, 20)

        # Ä°mleÃ§ Ã§evresinde mavi animasyon
        if self.hover_start_time:
            elapsed = time.time() - self.hover_start_time
            progress = int((elapsed / self.hover_duration) * 360)
            pen = QPen(QColor(0, 0, 255), 3)
            painter.setPen(pen)
            painter.drawArc(x - 30, y - 30, 60, 60, 0, progress * 16)

        painter.end()
        self.interface_label.setPixmap(pixmap)

    def keyPressEvent(self, event):
        if event.key() == Qt.Key_Q:
            self.close()

    def closeEvent(self, event):
        self.cap.release()
        self.face_mesh.close()
        event.accept()

    def open_gpt_grid_app(self, prompt_text):
        self.timer.stop()  # Timer'Ä± durdur
        self.cap.release()  # KamerayÄ± serbest bÄ±rak
        self.hide()  # Mevcut arayÃ¼zÃ¼ gizle
        self.gpt_grid_app = GPTGridApp(prompt_text)  # Yeni arayÃ¼zÃ¼ baÅŸlat
        self.gpt_grid_app.show()  # Yeni arayÃ¼zÃ¼ ekrana getir



class GPTGridApp(QWidget):
    def __init__(self, prompt_text, parent=None):
        super().__init__(parent)
        self.setWindowTitle("ChatGPT Interface")
        self.setGeometry(100, 100, 800, 600)

        # KaydÄ±rÄ±labilir baÅŸlÄ±k alanÄ±
        self.main_scroll_area = QScrollArea(self)
        self.main_scroll_area.setGeometry(50, 50, 700, 150)
        self.main_scroll_area.setStyleSheet("background-color: lightblue; border: 1px solid black;")
        self.main_scroll_area.setWidgetResizable(True)

        # BaÅŸlÄ±k iÃ§eriÄŸi iÃ§in QLabel
        self.main_label = QLabel(self.main_scroll_area)
        self.main_label.setStyleSheet("padding: 10px; color: black;")
        self.main_label.setFont(QFont("Arial", 14))
        self.main_label.setAlignment(Qt.AlignTop | Qt.AlignLeft)
        self.main_label.setWordWrap(True)
        self.main_scroll_area.setWidget(self.main_label)

        # SeÃ§enek alanlarÄ± iÃ§in liste
        self.option_areas = []  # SeÃ§enek butonlarÄ± ve alanlarÄ± iÃ§in liste
        self.create_option_areas()

        # Geri butonu
        self.back_button = QPushButton("Back", self)
        self.back_button.setGeometry(350, 520, 100, 40)
        self.back_button.setStyleSheet("background-color: lightgray; border: 1px solid black;")
        self.back_button.clicked.connect(self.go_back)
        self.back_button.setVisible(False)  # BaÅŸlangÄ±Ã§ta gÃ¶rÃ¼nmez

        # Ä°mleÃ§ pozisyonu ve hareketi iÃ§in Ã¶zellikler
        self.cursor_x = 400  # BaÅŸlangÄ±Ã§ pozisyonu
        self.cursor_y = 150
        self.hover_start_time = None
        self.hover_duration = 2  # Ä°mleÃ§ tÄ±klama sÃ¼resi
        self.cursor_radius = 20  # Ä°mleÃ§ boyutu

        # Ä°mleÃ§ koordinatlarÄ± QLabel'i
        self.cursor_label = QLabel(self)
        self.cursor_label.setGeometry(10, 10, 200, 30)  # EkranÄ±n sol Ã¼st kÃ¶ÅŸesi
        self.cursor_label.setStyleSheet("font-size: 14px; color: green;")
        self.cursor_label.setText("Coordinates: X: 0, Y: 0")

        self.cursor_animation_size = 10  # BaÅŸlangÄ±Ã§ boyutu
        self.animation_timer = QTimer(self)
        self.animation_timer.timeout.connect(self.animate_cursor)
        self.animation_timer.start(50)  # 50ms'de bir gÃ¼ncelle

        # Kamera Ã¶zellikleri
        self.timer = QTimer(self)
        self.timer.timeout.connect(self.update_frame)
        self.cap = cv2.VideoCapture(0)
        self.face_mesh = mp_face_mesh.FaceMesh(refine_landmarks=True, min_detection_confidence=0.7, min_tracking_confidence=0.7)

        # Hassasiyet ayarlarÄ±
        self.x_sensitivity = 6
        self.y_sensitivity = 10

        self.timer.start(20)

        # GPT Ã§aÄŸrÄ±sÄ±nÄ± tetikleyerek arayÃ¼zÃ¼ baÅŸlat
        self.history = []  # Ã–nceki adÄ±mlarÄ± saklamak iÃ§in
        self.fetch_gpt_response(prompt_text)

         # KaydÄ±rma butonlarÄ± iÃ§in tanÄ±mlar
        self.scroll_buttons = {
            "left": {"x": 10, "y": 100, "width": 40, "height": 40, "color": QColor(200, 200, 200), "action": self.scroll_left},
            "right": {"x": 750, "y": 100, "width": 40, "height": 40, "color": QColor(200, 200, 200), "action": self.scroll_right},
        }


        self.back_button.raise_()

    def create_option_areas(self):
        """
        SeÃ§enekler iÃ§in kaydÄ±rÄ±labilir alanlar ve altÄ±na butonlar oluÅŸturur.
        """
        for i in range(6):
            scroll_area = QScrollArea(self)
            scroll_area.setGeometry(50 + (i % 3) * 250, 250 + (i // 3) * 150, 200, 80)
            scroll_area.setStyleSheet("background-color: white; border: 1px solid black;")
            scroll_area.setWidgetResizable(True)

            label = QLabel(scroll_area)
            label.setStyleSheet("padding: 10px; color: black;")
            label.setFont(QFont("Arial", 12))
            label.setAlignment(Qt.AlignTop | Qt.AlignLeft)
            label.setWordWrap(True)
            scroll_area.setWidget(label)

            button = QPushButton("Select", self)
            button.setGeometry(50 + (i % 3) * 250, 340 + (i // 3) * 150, 200, 30)
            button.setStyleSheet("background-color: lightgray; border: 1px solid black;")
            button.setFont(QFont("Arial", 10))
            button.clicked.connect(self.on_button_click)

            self.option_areas.append((label, button))

    def update_frame(self):
        ret, frame = self.cap.read()
        if not ret:
            return

        # GÃ¶rÃ¼ntÃ¼yÃ¼ yatay eksende Ã§evir
        frame = cv2.flip(frame, 1)
        rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

        # Mediapipe ile yÃ¼z analizi
        results = self.face_mesh.process(rgb_frame)

        if results.multi_face_landmarks:
            for face_landmarks in results.multi_face_landmarks:
                # Burun ucu koordinatlarÄ±nÄ± al
                nose_tip = face_landmarks.landmark[1]
                new_cursor_x = int(nose_tip.x * self.width())
                new_cursor_y = int(nose_tip.y * self.height())

                # Hassas hareket iÃ§in koordinatlarÄ± yumuÅŸat
                self.cursor_x += int((new_cursor_x - self.cursor_x) / self.x_sensitivity)
                self.cursor_y += int((new_cursor_y - self.cursor_y) / self.y_sensitivity)

                # Ä°mlecin sÄ±nÄ±rlarÄ±nÄ± kontrol et
                self.cursor_x = max(0, min(self.cursor_x, self.width()))
                self.cursor_y = max(0, min(self.cursor_y, self.height()))

                # Scroll alanÄ±nda hareket algÄ±la ve kaydÄ±rma yap
                if self.main_scroll_area.geometry().contains(self.cursor_x, self.cursor_y):
                    scroll_bar = self.main_scroll_area.verticalScrollBar()
                    if self.cursor_y < self.main_scroll_area.y() + 50:  # YukarÄ± kaydÄ±rma
                        scroll_bar.setValue(scroll_bar.value() - 10)
                    elif self.cursor_y > self.main_scroll_area.y() + self.main_scroll_area.height() - 50:  # AÅŸaÄŸÄ± kaydÄ±rma
                        scroll_bar.setValue(scroll_bar.value() + 10)

                # Hover iÅŸlemini kontrol et
                self.check_hover()

                # Ä°mleÃ§ koordinatlarÄ±nÄ± QLabel'e yaz
                self.cursor_label.setText(f"Coordinates: X: {self.cursor_x}, Y: {self.cursor_y}")

        # Kamera gÃ¶rÃ¼ntÃ¼sÃ¼nÃ¼ QLabel'e aktar
        h, w, ch = rgb_frame.shape
        bytes_per_line = ch * w
        
        # EkranÄ± yeniden Ã§iz
        self.update()

    def animate_cursor(self):
        # Ä°mlecin boyutunu titreÅŸim etkisi iÃ§in deÄŸiÅŸtir
        self.cursor_animation_size += 1
        if self.cursor_animation_size > 15:
            self.cursor_animation_size = 10
        self.update()

    def check_hover(self):
        # Geri butonu kontrolÃ¼
        if self.back_button.isVisible() and self.back_button.geometry().contains(self.cursor_x, self.cursor_y):
            if self.hover_start_time is None:
                self.hover_start_time = time.time()
            elif time.time() - self.hover_start_time >= self.hover_duration:
                self.go_back()
                self.hover_start_time = None
            return

        # SeÃ§enek butonlarÄ± kontrolÃ¼
        for _, button in self.option_areas:
            if button.geometry().contains(self.cursor_x, self.cursor_y):
                if self.hover_start_time is None:
                    self.hover_start_time = time.time()
                elif time.time() - self.hover_start_time >= self.hover_duration:
                    button.click()
                    self.hover_start_time = None
                return

        # KaydÄ±rma butonlarÄ± kontrolÃ¼
        for button_key, button in self.scroll_buttons.items():
            if self.is_cursor_on_button(button):
                if self.hover_start_time is None:
                    self.hover_start_time = time.time()
                elif time.time() - self.hover_start_time >= self.hover_duration:
                    button["action"]()  # KaydÄ±rma iÅŸlemini gerÃ§ekleÅŸtir
                    self.hover_start_time = None
                return

        # Hover baÅŸlangÄ±cÄ±nÄ± sÄ±fÄ±rla
        self.hover_start_time = None

    def fetch_gpt_response(self, prompt_text):
        """
        GPT'den bir yanÄ±t alÄ±r ve arayÃ¼zÃ¼ gÃ¼nceller.
        """
        try:
            self.initial_query = prompt_text  # KullanÄ±cÄ±nÄ±n sorgusunu sakla
            response = openai.ChatCompletion.create(
                model="gpt-3.5-turbo",
                messages=[
                    {"role": "system", "content": "You are an AI assistant. Provide a detailed 'full sentence' response, "
                                                "and generate 6 concise options as a numbered list for the buttons."},
                    {"role": "user", "content": prompt_text}
                ]
            )
            full_text = response['choices'][0]['message']['content']

            lines = full_text.split("\n")
            detailed_response = []
            numbered_options = []

            for line in lines:
                if line.strip().startswith(("1.", "2.", "3.", "4.", "5.", "6.")):
                    numbered_options.append(line.strip())
                else:
                    detailed_response.append(line.strip())

            # YanÄ±tÄ± kaydet
            self.initial_response = full_text  # Tam yanÄ±tÄ± sakla
            self.full_response = "\n".join(detailed_response)  # DetaylÄ± cevabÄ± sakla
            self.main_label.setText(self.full_response)
            self.options = numbered_options
            self.update_option_buttons(self.options)

            self.history.append((self.full_response, self.options))
            
            # ArayÃ¼zÃ¼ yeniden Ã§iz
            self.update()

        except Exception as e:
            self.main_label.setText(f"Hata: {str(e)}")

    def paintEvent(self, event):
        """
        Ä°mleci ekran Ã¼zerine Ã§izer.
        """
        # Ã–nce ana widget Ã§izimini yap
        super().paintEvent(event)

        # Daha sonra imleci Ã§iz
        painter = QPainter(self)
        painter.setBrush(QColor(255, 0, 0))  # KÄ±rmÄ±zÄ± renk
        painter.setPen(Qt.NoPen)  # Ã‡izgi yok, sadece dolgu
        painter.drawEllipse(QPoint(self.cursor_x, self.cursor_y), self.cursor_radius, self.cursor_radius)

        # SeÃ§enek Ã¼zerindeki mavi animasyon
        if self.hover_start_time:
            elapsed = time.time() - self.hover_start_time
            progress = int((elapsed / self.hover_duration) * 360)
            pen = QPen(QColor(255, 165, 0), 3)
            painter.setPen(pen)
            painter.drawArc(self.cursor_x - 30, self.cursor_y - 30, 60, 60, 0, progress * 16)

        painter.end()

    def update_option_buttons(self, options):
        for i, (label, button) in enumerate(self.option_areas):
            if i < len(options):
                option_text = options[i]
                label.setText(option_text)
                label.parentWidget().setVisible(True)
                button.setVisible(True)
                button.setEnabled(True)
            else:
                label.setText("")
                label.parentWidget().setVisible(False)
                button.setVisible(False)
                button.setEnabled(False)

    def on_button_click(self):
        clicked_button = self.sender()
        index = next((i for i, (_, btn) in enumerate(self.option_areas) if btn == clicked_button), -1)
        if index != -1 and index < len(self.options):
            selected_option = self.options[index]
            self.history.append((self.full_response, self.options))  # Mevcut durumu kaydet
            self.fetch_subcategories(selected_option)

    def fetch_subcategories(self, selected_option):
        """
        Fetches detailed explanation and subcategories for the selected option.
        """
        try:
            response = openai.ChatCompletion.create(
                model="gpt-3.5-turbo",
                messages=[
                    {"role": "system", "content": "You are an AI assistant. Ensure your responses are connected to the initial query and answer."},
                    {"role": "user", "content": f"Initial query: {self.initial_query}"},
                    {"role": "assistant", "content": f"Initial response: {self.initial_response}"},
                    {"role": "user", "content": f"Now, generate a detailed explanation and subcategories for: {selected_option}"}
                ]
            )
            full_text = response['choices'][0]['message']['content']

            lines = full_text.split("\n")
            detailed_description = []
            numbered_subcategories = []

            for line in lines:
                if line.strip().startswith(("1.", "2.", "3.", "4.", "5.", "6.")):
                    numbered_subcategories.append(line.strip())
                else:
                    detailed_description.append(line.strip())

            if self.full_response or self.options:
                self.history.append((self.full_response, self.options)) 

            self.full_response = "\n".join(detailed_description)
            self.main_label.setText(self.full_response)
            self.options = numbered_subcategories
            self.update_option_buttons(self.options)

            # Geri butonunu gÃ¶rÃ¼nÃ¼r yap
            self.back_button.setVisible(True)

            print(f"Geri butonunun geometrisi: {self.back_button.geometry()}")

        except Exception as e:
            self.main_label.setText(f"Hata: {str(e)}")

    
    def repaint_interface_label(self):
        pixmap = QPixmap(self.main_scroll_area.size())
        pixmap.fill(Qt.transparent)

        painter = QPainter(pixmap)

        # KaydÄ±rma butonlarÄ±nÄ± Ã§iz
        for button in self.scroll_buttons.values():
            painter.setBrush(button["color"])
            painter.drawRect(button["x"], button["y"], button["width"], button["height"])

        # Ä°mleÃ§ Ã§iz
        painter.setBrush(QColor(255, 0, 0))
        painter.setPen(Qt.NoPen)
        painter.drawEllipse(QPoint(self.cursor_x, self.cursor_y), self.cursor_radius, self.cursor_radius)

        painter.end()
        self.main_scroll_area.setPixmap(pixmap)

    def scroll_left(self):
        self.main_scroll_area.horizontalScrollBar().setValue(self.main_scroll_area.horizontalScrollBar().value() - 20)

    def scroll_right(self):
        self.main_scroll_area.horizontalScrollBar().setValue(self.main_scroll_area.horizontalScrollBar().value() + 20)

    def is_cursor_on_button(self, button):
        return (button["x"] <= self.cursor_x <= button["x"] + button["width"] and
                button["y"] <= self.cursor_y <= button["y"] + button["height"])

    def go_back(self):
        if self.history:
            # GeÃ§miÅŸten Ã¶nceki durumu al
            previous_full_response, previous_options = self.history.pop()
            self.full_response = previous_full_response
            self.options = previous_options

            # Ana baÅŸlÄ±k ve seÃ§enekleri geri yÃ¼kle
            self.main_label.setText(self.full_response)
            self.update_option_buttons(self.options)

            # Geri butonunu gizle (geÃ§miÅŸ boÅŸ ise)
            self.back_button.setVisible(len(self.history) > 0)
        else:
            self.main_label.setText("GeÃ§miÅŸ bulunamadÄ±!")  # Test iÃ§in mesaj

    def keyPressEvent(self, event):
        if event.key() == Qt.Key_Q:  # Q tuÅŸuna basÄ±ldÄ±ÄŸÄ±nda
            self.close()  # UygulamayÄ± kapat

if __name__ == "__main__":
    app = QApplication(sys.argv)
    camera_feed = CameraFeed()
    camera_feed.show()
    sys.exit(app.exec_())
```
